```python
import pandas as pd
from datetime import datetime


df['registration_date'] = pd.to_datetime(df['registration_date'], dayfirst=True, errors='coerce')

# Add history tracking columns
today = pd.to_datetime(datetime.today().date())
df['start_date'] = today
df['end_date'] = pd.NaT

# Simulate a change: update email for customer_id = 2
current = df[(df['customer_id'] == 2) & (df['end_date'].isna())]

if not current.empty and current.iloc[0]['email'] != 'pavani@gmail.com':
    # Expire the old version
    df.loc[(df['customer_id'] == 2) & (df['end_date'].isna()), 'end_date'] = today

    # Create new version row
    new_row = current.copy()
    new_row.loc[:, 'email'] = 'pavani@gmail.com'
    new_row.loc[:, 'start_date'] = today
    new_row.loc[:, 'end_date'] = pd.NaT

    # Append and sort both rows together
    df = pd.concat([df, new_row], ignore_index=True)

# ✅ Sort to get old + new together for each customer
df = df.sort_values(by=['customer_id', 'start_date']).reset_index(drop=True)

# ✅ Display result for customer_id = 2
df[df['customer_id'] == 2]
```

```python
import pandas as pd
from datetime import datetime

# ✅ 1. Use your already loaded df as main table
df_main = df.copy()

# ✅ 2. Create or initialize history table (Type 4 part)
df_history = pd.DataFrame(columns=df_main.columns.tolist() + ['changed_on'])

# ✅ 3. Define customer update
customer_id = 2
new_email = 'pavani@gmail.com'
mask = df_main['customer_id'] == customer_id

# ✅ 4. If customer exists and email is changing
if not df_main.loc[mask].empty and df_main.loc[mask, 'email'].iloc[0] != new_email:
    
    # 🔁 a) Log current version into history table (SCD Type 4)
    old_record = df_main.loc[mask].copy()
    old_record['changed_on'] = pd.to_datetime(datetime.today().date())
    df_history = pd.concat([df_history, old_record], ignore_index=True)

    # 🔁 b) Overwrite main table with new email (SCD Type 1)
    df_main.loc[mask, 'email'] = new_email

# ✅ 5. Output both tables
print("🔵 Main Table (Current Values):")
print(df_main[df_main['customer_id'] == customer_id])

print("\n🟡 History Table (All Old Versions):")
print(df_history[df_history['customer_id'] == customer_id])
```
