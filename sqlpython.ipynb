```python
import pandas as pd
```

```python
df1=pd.read_csv("transaction_data.csv")
df1
```

```python
df2=pd.read_csv("us_customer_data1.csv")
df2
```

```python
a=pd.merge(df1,df2,on="customer_id",how="inner")
a
```

```python
a.duplicated()
```

```python
a.isnull().sum()
```

```python
a["amount"]=a["amount"].fillna(0)
a["email"]=a["email"].fillna("example.@gmail.com")
a["phone"]=a["phone"].fillna("XXX-XXX-XXXX")
a
```

```python
a['email']=a['email'].fillna(a['name'].str.lower().replace(" ",".")+"@gmail.com")
```

```python
a.head(100)
```

```python
a['phone']=a['phone'].str.replace(r'[^0-9]',"",regex=True)
```

```python
a
```

```python
a["country_code"]=a["address"].str[-8:-6]
```

```python
a
```

```python
b=pd.read_csv("cleaned_countrycodes.csv")
```

```python
c=pd.merge(a,b,on="country_code",how="left")
```

```python
c
```

```python
c["CallingCode"]=c["CallingCode"].fillna(1)
```

```python
c
```

```python
c["phone"]="+"+c["CallingCode"].astype(str)+" "+c["phone"].astype(str)
```

```python
c
```

```python
c=c[c["amount"]>c["amount"].mean()]
```

```python
c
```

```python
c.to_csv("cleaned.csv",index=False)
```
